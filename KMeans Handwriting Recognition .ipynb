# KMEANS HANDWRITING RECOGNITION: SARWAAN ANSARI

Importing Numpy, matplotlib, sklearn (for digit dataset), and KMeans (the algorithm we are using)

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
```

Get digit data and print the data. The digits are a 8x8 image of each digit.

```python
digits = datasets.load_digits()
print(digits.data)
print(digits.target)
```

```python
plt.gray()
plt.matshow(digits.images[100])
plt.show()
print(digits.target[100])
```

We use k = 10 becuase we only have 10 digits.

```python
#10 digits:
k = 10
model = KMeans(n_clusters=k, random_state = 40)
model.fit(digits.data)
```

Plot the different cluster centers to describe each digit.

```python
fig = plt.figure(figsize = (8, 3))
fig.suptitle("Images", fontsize = 14, fontweight = 'bold')

for i in range(10):
  ax = fig.add_subplot(2, 5, 1 + i)
  ax.imshow(model.cluster_centers_[i].reshape((8, 8)), cmap=plt.cm.binary)

plt.show()
```
